```python
import pandas as pd  
import numpy as np 
```

```python
url='https://raw.githubusercontent.com/stanleyokali/files/main/food.csv'
df = pd.read_csv(url,encoding ='cp1252')
df
```

**What is the total sum of Animal Fat produced in 2014 and 2017 respectively?**

**Which year had the highest correlation with ‘Element Code’?**

```python
a=df['Element Code'].corr(df['Y2014'])
b=df['Element Code'].corr(df['Y2015'])
c=df['Element Code'].corr(df['Y2016'])
d=df['Element Code'].corr(df['Y2017'])
e=df['Element Code'].corr(df['Y2018'])
```

```python
print(a,b,c,d,e)

#a=2014 has the highest corellation with Element Code
```

**Which of these elements had the highest sum in 2018? a.import quantity b.domestic supply quantity c. food supply quantity d.food supply**

```python
#Hint-  Select columns ‘Y2018’ and ‘Element’,
#Perform a groupby operation on ‘Element’ 
#on the selected dataframe and answer the  question.

df.groupby('Element')['Y2018'].sum()
xx=df.groupby('Element')['Y2018'].sum()
xx
sorted = xx.sort_values(ascending=False)
print(sorted.head())


#answer = b (domestic supply quantity)
```

**Which of these elements had the 3rd lowest sum in 2018? a.fat supply quantity b.seed c.food supply quantity d.food supply e.protein supply quantity**

```python
#Hint-  Select columns ‘Y2018’ and ‘Element’, 
#Perform a groupby operation on ‘Element’ on the selected dataframe and answer the  question. 

df.groupby('Element')['Y2018'].sum()
xx=df.groupby('Element')['Y2018'].sum()
xx
sorted = xx.sort_values()
print(sorted.head())


#answer is e = protein supply quantity
```
